```python
sheet = wb.worksheet('Testing Output')
data_testing = sheet.get_all_values()
y_test = pd.DataFrame(data_testing)
y_test.columns = y_test.iloc[0]
y_test = y_test.iloc[1:]
y_test
```

# Decision tree

```python
from sklearn.metrics import r2_score, mean_squared_error
from  math import sqrt
```

```python
# Import the necessary modules and libraries
import numpy as np
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt

print('Max dept  R2_train               R2_test                  RMSE_train             RMSE_test')
for i in range(1,7):
    regr = DecisionTreeRegressor(max_depth=i)
    regr.fit(X_train, y_train)
    yp_test = regr.predict(X_test)
    yp_train = regr.predict(X_train)

    r2_train = r2_score(y_train, yp_train)
    r2_test = r2_score(y_test, yp_test)
    rmse_train = sqrt(mean_squared_error(y_train,yp_train)) #calculate rmse
    rmse_test = sqrt(mean_squared_error(y_test,yp_test)) #calculate rmse
    print('   ',i , '   ',r2_train, '   ',r2_test,   '    ', rmse_train, '    ', rmse_test)
```
